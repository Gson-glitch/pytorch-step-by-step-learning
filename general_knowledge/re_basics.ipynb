```python
import re
```

# General `re` syntax: re.func(pattern, sequence)

## .match()

```python
# Finds only the first occurrence of a `pattern` at the beginning of a `sequence`.
# Basically answers: does this sequence start with the pattern?
# group(0) same as group() 0 is the default returns the matched word if it exists. 
# If not accessing .group() throws an error: 'NoneType' object has no attribute 'group'
found = re.match(r"hello", "hello, hello world")
found, found.re.groups, found.group()
```

```python
# Returns None because the sequence doesn't start with hello.
re.match(r"hello", "hey. hello, world")
```

## .search()

```python
# Answers the question, is this pattern anywhere in the sequence? If true, return only the first occurrence
found = re.search(r"hello", "hey, hello, hello world")
found, found.group()
```

```python
found = re.search(r"(hello)", "hey, hello, hello world")
found, found.group()
```

## .findall()

```python
# Answers the question how many occurrences of `pattern` exist in the `sequence`
# This is a list object so it doesn't have the .group() property
found = re.findall(r"hello", "hey, hello, hello world")
found
```
