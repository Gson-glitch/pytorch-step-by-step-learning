```python
import matplotlib.pyplot as plt
```

```python
values = [81.3, 80.3, 75.3, 80.3, 85.2, 90.2, 95.2, 90.2, 89.1, 88.8]
plt.plot(values, marker='x')
plt.grid()
plt.show()
```

```python
from typing import List

def count_peaks(values: List[float]) -> int:
    peaks = 0
    n = len(values)

    for i in range(1, n - 1):
        left = values[i - 1]
        mid = values[i]
        right = values[i + 1]

        # Top peak
        if mid >= left + 5 and mid >= right + 5:
            peaks += 1
        # Bottom peak
        elif mid <= left - 5 and mid <= right - 5:
            peaks += 1

    return peaks
```

```python
count_peaks(values=values)
```
