````python
import numpy as np
````

Think of it like this, if you give meshgrid 2 arrays: x and y, you're asking: give me all (x, y) arranged in a grid form. 

Consider the example below, if I wrote the pairs myself, I'd have: 

```python
[(1, 10), (2, 10), (3, 10),
 (1, 20), (2, 20), (3, 30)]


````python
x = [1, 2, 3]
y = [10, 20]

X, Y = np.meshgrid(x, y)
np.meshgrid(x, y)
````

````python
pairs = X.flatten(), Y.flatten()
pairs
````

````python
np.stack(pairs, axis=1)
````
